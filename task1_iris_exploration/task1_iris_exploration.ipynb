# Task 1: Iris Dataset Exploration
Load, inspect, and visualize the Iris dataset with pandas, seaborn, and matplotlib.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')
print('Shape:', df.shape)
print('Columns:', list(df.columns))
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.show()
df.hist(figsize=(10, 7), bins=15)
plt.show()
melted = df.melt(id_vars='species', var_name='feature', value_name='value')
sns.boxplot(data=melted, x='feature', y='value', hue='species')
plt.xticks(rotation=15)
plt.show()
```
