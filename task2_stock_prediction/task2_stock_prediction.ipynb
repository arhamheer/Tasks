# Task 2: Short-Term Stock Prediction
Use yfinance and regression to predict the next day's close price.

```python
import yfinance as yf
import pandas as pd
from sklearn.linear_model import LinearRegression

df = yf.download('AAPL', period='2y', progress=False).reset_index()
df['TargetClose'] = df['Close'].shift(-1)
df = df.dropna()
df.head()
```

```python
X = df[['Open', 'High', 'Low', 'Volume']]
y = df['TargetClose']
split = int(len(df) * 0.8)
model = LinearRegression().fit(X.iloc[:split], y.iloc[:split])
pred = model.predict(X.iloc[split:])
pred[:5]
```
